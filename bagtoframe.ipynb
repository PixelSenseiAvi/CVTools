```python
import cv2
import pyrealsense2 as prs
import os
import numpy as np
```

```python
fft_path = 'D:/surgical_knot_tying/surgical_knot_tying'
save_path = 'D:/surgical_knot_tying/frames'
    
i = 0
for foldername in os.listdir(fft_path):
    if os.path.isdir(os.path.join(fft_path, foldername)):
        for filename in os.listdir(os.path.join(fft_path, foldername)):
            if os.path.isfile(os.path.join(fft_path, foldername, filename)):

                if not os.path.exists(os.path.join(save_path, foldername)):
                    os.mkdir(os.path.join(save_path, foldername))
                if not os.path.exists(os.path.join(save_path, foldername, filename[:-4])):
                    os.mkdir(os.path.join(save_path, foldername, filename[:-4]))
                
                i+=1
                pipeline = prs.pipeline()
                    # create a config object
                config = prs.config()

                prs.config.enable_device_from_file(config, os.path.join(fft_path, foldername, filename))
                config.enable_stream(prs.stream.color, 640, 480, prs.format.rgb8, 15)

                fht_path = os.path.join(os.path.join(save_path, foldername) + "\\"+
                                        filename[:-4], str(i), '.avi')
                #fourcc = cv2.VideoWriter_fourcc(*'XVID')
                #out = cv2.VideoWriter(fht_path,fourcc, 20.0, (640,480))
                profile = pipeline.start(config)

                j=0
                try:
                    while True:
                        
                        j+=1
                        frames = pipeline.wait_for_frames()
                        
                        color_frame = frames.get_color_frame()
                        color_image = np.asanyarray(color_frame.get_data())
                        color_image = cv2.cvtColor(color_image, cv2.COLOR_BGR2RGB)

                        cv2.namedWindow('Align Example', cv2.WINDOW_AUTOSIZE)
                        print(os.path.join(save_path, foldername, filename[:-4],
                                           str(j)+'.jpg'))
                        cv2.imwrite(os.path.join(save_path, foldername, filename[:-4],
                                     str(j)+'.jpg'),color_image)
                        #out.write(color_image)
                        cv2.imshow('Align Example', color_image)

                        key = cv2.waitKey(1)
                            # Press esc or 'q' to close the image window
                        if key & 0xFF == ord('q') or key == 27:
                            cv2.destroyAllWindows()
                            break

                finally:
                    pipeline.stop()
```
